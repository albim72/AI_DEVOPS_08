```python
#jak wygląda wykres funkcji sigmoid?

import matplotlib.pyplot as plt
import numpy as np

# Defining the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Creating an array of values between -12 and 12
x = np.linspace(-12, 12, 500)

# Plotting
plt.figure(figsize=(9,6))
plt.grid(True)
plt.plot(x, sigmoid(x), 'g') # 'g' specifies that the line color should be green
plt.title('Sigmoid Function')
plt.show()
```
